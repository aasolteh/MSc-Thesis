```python
features = os.listdir(accuracyPath)
```

```python
probeDict = {
    "VISp"  : ["PrimaryVisualCortex (V1)"],
    "VISl"  : ["LateroMedial (LM)"],
    "VISrl" : ["RostroLateral (RL)"],
    "VISal" : ["AnteroLateral (AL)"],
    "VISpm" : ["PosteroMedial (PM)"],
    "VISam" : ["AnteroMedial (AM)"],
    "LP"    : ["LateralPosteriorNuc (LP)"],
    "LGd"   : ["LateralGeniculateNuc (LGn)"]
}
```

```python
sortedProbes = ['LGn', 'V1', 'LM', 'AL', 'PM', 'RL', 'AM']
```

```python
def find_peak(data):
    return np.max(data)
```

```python
for type in features:
    featurePath = os.path.join(accuracyPath, type)
    probesData  = os.listdir(featurePath)
    probeNames  = []
    for probe in probesData:
        print(probe)
        probeName = probe.split("(")[-1].split(")")[0]
        if probeName == "LP":
            continue
        probePath = os.path.join(featurePath, probe)
        probeData = np.mean(np.load(probePath), axis=0)
        peak = find_peak(probeData)
        print(peak)
        print(probeData.shape)
        probeNames.append(probeName)
        
    break
```
