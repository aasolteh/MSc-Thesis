```python
import os
from allensdk.brain_observatory.ecephys.ecephys_project_cache import EcephysProjectCache
import numpy as np
import pandas as pd
```

```python
data_directory = 'raw_data/'
manifest_path = os.path.join(data_directory, "manifest.json")
cache = EcephysProjectCache.from_warehouse(manifest=manifest_path)

sessions = cache.get_session_table()
```

```python
gazeDataPath = '/home/amirali/Desktop/Thesis/Codes/gaze_data/'
```

```python
for sessIdx in range(len(sessions.index.values)):
    print("Session: ", sessIdx)
    currSessNum = sessions.index.values[sessIdx]
    currSession = cache.get_session_data(currSessNum)

    if not 'natural_scenes' in currSession.stimulus_names:
        continue
    
    gaze_data = currSession.get_screen_gaze_data()
    if gaze_data is None:
        continue
    
    gaze_data.to_csv(path_or_buf=gazeDataPath + str(currSessNum) + '.csv', index = True)
```
